```python
import pandas as pd

df = pd.read_csv(
    '/kaggle/input/datasets/muhammadumer7804/customer-churn-dataset/Telco-Customer-Churn.csv'
)

df.head()
```

```python
df.shape
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())

total_customers = len(df)

churn_rate = round(
    df['Churn'].value_counts(normalize=True)['Yes'] * 100,
    2
)

avg_monthly_charges = round(df['MonthlyCharges'].mean(), 2)

avg_tenure = round(df['tenure'].mean(), 2)

print("Total Customers:", total_customers)
print("Churn Rate:", churn_rate, "%")
print("Average Monthly Charges:", avg_monthly_charges)
print("Average Tenure:", avg_tenure)
```
